```python
# OBAT CHARLES PROJECTS
#Diabetes Prediction using Machine Learning with Python
#Importing the required libraries
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn import svm
#Loading the dataset
DiabetesData=pd.read_csv('diabetes.csv')
#Viewing the first five rows of the dataset
DiabetesData.head()
#Classification of the outcome column
#print(DiabetesData['Outcome'].unique())
#Checking how many 1 and 0 are there in the outcome column ie how many ppl are considered to be diabetic and non diabetic based on the dataset
#print(DiabetesData['Outcome'].value_counts())
#print(DiabetesData.columns)
#DiabetesData.info()#Checking for the datatype of each column
#Now lets try to group the dataset based on the label
#print(DiabetesData.groupby('Outcome').mean())
#DiabetesData.groupby('Outcome').mean().plot(kind='bar',figsize=(14,6))
#Checking  for Outliers
plt.figure(figsize=(14,6))
#DiabetesData.boxplot(figsize=(14,6))#Ouliers Present in most of the dataset columns

#We can now separate the data into features and lables
X=DiabetesData.drop(columns='Outcome',axis=1)
Y=DiabetesData['Outcome']
#Remove Outliers using for of loop and where()medhod from numpy
for column in X.columns:
  q1=X[column].quantile(0.25)
  q3=X[column].quantile(0.75)
  IQR=q1-q3
  Upper_Limit=q3+1.5*IQR
  Lower_Limit=q1-1.5*IQR
X[column]=np.where(X[column]>Upper_Limit,Upper_Limit,np.where(X[column]<Lower_Limit,Lower_Limit,X[column]))
#Checking for Outliers Once More
plt.figure(figsize=(14,6))
#DiabetesData.boxplot(figsize=(14,6))#Ouliers Absent in most of the dataset columns
#Checking for missing value
DiabetesData.isnull().mean()#No missng value present
#Using StandardScalar Class to scale the features dataset
scaler=StandardScaler()
X_Scaled=scaler.fit(X).transform(X)#Returns a numpy Array that needs to be converted to a pandas dataframe
#print(X_Scaled)#numpy Array
X=pd.DataFrame(data=X_Scaled,columns=X.columns)
#print(X.head(),X.describe())
#Spliting the Data into train and test data
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,train_size=0.9,random_state=2,stratify=Y)
print(X_test.head())
print(Y_test.head())
#Training the model
model_classifier=svm.SVC(kernel='linear')
model_classifier.fit(X_train,Y_train)
#Making Prediction on train data
X_train_prediction=model_classifier.predict(X_train)
#Model Evaluation on train data
print(accuracy_score(Y_train,X_train_prediction))#Our model has an accuracy of 77.71345875542692%
#Making Prediction on test data
X_test_prediction=model_classifier.predict(X_test)
#Model Evaluation on test data
print(accuracy_score(Y_test,X_test_prediction))#Our model has an accuracy of 77.92207792207793%
# actual_prediction=pd.DataFrame(data=[Y_test.values,Y_prediction],index=['Actual_Value','Predicted_Value'])
# print(actual_prediction)
actual_Y=pd.DataFrame(data=Y_test.values,columns=['Actual_Value'])
X_test_prediction=pd.DataFrame(data=X_test_prediction,columns=['Predicted_Value'])
merged_Y=pd.concat([actual_Y,X_test_prediction],axis=1)
print(merged_Y.head(70))
#Making prediction on a single record/input data i.e making a system for prediction
input_data=(X.values[11].reshape(1,-1))

print('data:',input_data)
predicted_value=model_classifier.predict(input_data)
print('Predicted Value:',predicted_value)#Returns a list
print('Actual Value:',Y[11])
if predicted_value[0]==0:
  print('The person is not diabetic')
else:
  print('The person is diabetic')
```
