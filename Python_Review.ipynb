```python
# Data preprocessing- involves cleaning and engineering data in a
# way that it can be used as input to several important data science
# tasks, such as data visualization, machine learning, deep learning,
# and data analytics.
#  Some of the most common data preparation
# tasks include feature scaling, handling missing values, categorical
# variable encoding, data discretization, etc.
# Scikit-learn library contains various modules and libraries that you
# can employ to perform different types of data processing tasks
#Feature scaling is applied on numeric columns only.
# Standardization is the process of centering a variable at zero and
# standardizing the data variance to 1. To standardize a dataset, you
# simply have to subtract each data point from the mean of all the
# data points and divide the result by the standard deviation of the data.
# To standardize the data, you can use the StandardScaler class
# from the sklearn.preprocessing module. You have to pass the
# Pandas dataframe containing the dataset to the fit() method of the
# class and then to the transorm() method of the class. The following
# script applies standard scaling on the age, fare, and pclass columns
# of the Titanic dataset.
#the standard scaling doesnt actually affect the default data distribution.

import pandas as pd
import seaborn as sns
from sklearn.preprocessing import StandardScaler
data=sns.load_dataset('titanic')
data=data[['age','pclass','fare']]#Select only the numeric columns
print(data['pclass'].max())
scaled_Data=StandardScaler().fit(data).transform(data)
#print(scaled_Data)
#Now convert the scaled Data into a pandas dataframe with exact colums as those of the data variable
Pd_dataFrame=pd.DataFrame(data=scaled_Data,columns=data.columns)
print(Pd_dataFrame.describe())
print(Pd_dataFrame['pclass'].max())
sns.kdeplot(Pd_dataFrame['pclass'])

```

```python
#  Handling Missing Numerical Data
# One of the most commonly occurring data types is numeric data, which consists of numbers
# To handle missing numerical data, we can use statistical techniques.
#  The use of statistical techniques or algorithms to replace missing values with
#   statistically generated values is called imputation.
# Complete case analysis (CCA), also known as list wise deletion is the most basic technique
#  for handling missing data. In CCA you drop all rows with missing data and run your analysis
#  only on the remaining complete cases.

# Mean or Median Imputation
# In mean or median imputation, missing values in a column are
# replaced by the mean or median of all the remaining values in that particular column.
# Lets see a practical example of mean and median imputation. We
# will import the Titanic dataset and find the columns that contain
# missing values. Then, we will apply mean and median imputation to
# the columns containing missing values, and, finally, we will see the
# effect of applying mean and median imputation to the missing values.
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

titanicData=sns.load_dataset('titanic')
titanicData.columns
# Index(['survived', 'pclass', 'sex', 'age', 'sibsp', 'parch', 'fare',
#        'embarked', 'class', 'who', 'adult_male', 'deck', 'embark_town',
#        'alive', 'alone'],
#       dtype='object')
# Lets filter some of the numeric columns from the dataset and see if
# they contain any missing values.
titanicData=titanicData[['survived', 'pclass','age', 'fare']]
titanicData.head()
print(titanicData.describe())
# To find/see if aforementioned columns contain any  missing values you
# need to first call the isnull() method on the titanicData dataframe,
# and then, you need to call the mean() method as shown below
titanicData.isnull().mean()
# The output shows that only the age column contains missingvalues.
# And the ratio of missing values is around 19.86 percent.
# Lets now find out the median and mean values for all the nonmissing values in the age column.
meanAge=titanicData['age'].mean()
medianAge=titanicData['age'].median()
#Adding two columns ie(mean and median age columns) to the titanicData dataframe
titanicData['meanAge']=titanicData['age'].fillna(np.round(meanAge,1))
titanicData['medianAge']=titanicData['age'].fillna(medianAge)
titanicData.head(20)
print(titanicData.describe())
# You can clearly see that the default values in the age columns have
# been distorted by mean and median imputation, and the overall
# variance of the dataset has also been decreased.
#PLOTING THE Data
sns.set_style('darkgrid')
fig=plt.figure()
ax=fig.add_subplot(111)
titanicData['age'].plot(kind='kde',color='green',ax=ax)
titanicData['meanAge'].plot(kind='kde',color='blue',ax=ax)
titanicData['medianAge'].plot(kind='kde',color='pink',ax=ax)
lines,lables=ax.get_legend_handles_labels()
#print(lines,lables)
ax.legend(lines,lables,loc='best')



```

```python
#  Frequent Category Imputation
# One of the most common ways of handling missing values in a
# categorical column is to replace the missing values with the most
# frequently occurring values, i.e., the mode of the column.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['figure.figsize']=[10,10]
sns.set_style('darkgrid')#It changes the default aesthetics of your plots to use a dark grid background.
# Seaborn has five built-in themes:
# "darkgrid"  dark background with grid lines (default)
# "whitegrid"  white background with grid lines
# "dark"  dark background, no grid
# "white"  white background, no grid
# "ticks"  white background with axis ticks
fig=plt.figure()
xa=fig.add_subplot(1,1,1)
dataa=sns.load_dataset('titanic')
dataa=pd.DataFrame(dataa)
#dataa.plot(kind='kde')
dataa=dataa[['embarked','survived', 'pclass','age', 'fare']]
#dataa.isnull().mean()
#.mode() does not return a single value, it returns a Series (because there can be multiple modes).
#use [0] to Pick the first mode value explicitlyfrom the series
embarkedMode = dataa['embarked'].mode()[0]
ageMode = dataa['age'].mode()[0]
print(ageMode)
dataa = dataa.fillna({'embarked': embarkedMode, 'age': ageMode})
#add mAge columns in the dataframe
dataa['mAge']=dataa['age'].fillna(ageMode)
print(dataa.head())
dataa.isnull().mean()
#plotting the data of age distribution
dataa['mAge'].plot(kind='kde',color="red",ax=xa)
dataa['age'].plot(kind='kde',color="black",ax=xa)
lines,lables=xa.get_legend_handles_labels()
xa.legend(lines,lables,loc="best")
```

```python
## Categorical Data Encoding
# Models based on statistical algorithms, such as machine learning and deep learning, work with numbers/numerical data.
#In datasets, some features are numerical (age, salary, height, etc.)  while others are
#categorical (sex, country, color, yes/no, etc.).
# Machine learning models (like logistic regression, SVM, random forest, etc.) work with numbers, not text.
# So we need to convert categorical data  numerical values before training a model.
# This is called categorical encoding.
# Types of categorical data
# There are two main types:
#  Nominal data
# Categories dont have an order.
# Example: "red", "green", "blue" or "male", "female".
# Solution: One-Hot, Dummy, Frequency, Target Encoding.
#  Ordinal data
# Categories have a meaningful order.
# Example: "low < medium < high" or "small < medium < large".
# Solution: Ordinal or Label Encoding.

#One Hot Encoding-Creates a new column for each category with 0/1 values.
# One hot encoding is one of the most commonly used categorical encoding schemes. In one hot encoding,
#  for each unique value in the categorical column, a new column is added. Integer 1 is added to the column that
#  corresponds to original label, and all the remaining columns are filled with zeros.

# The easiest way to convert a column into one hot-encoded column
# is by using the get_dummies() method of the Pandas dataframe,
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
#pd.set_option('display.max_columns', None)  # None = no column limit
#Load titatinic dataset from seaborn
titanicData=sns.load_dataset('titanic')
#titanicData.info() #see the datatype of each column
#Lets filter the titanic_data into two ie numeric and categorical data
NumeriDAta=titanicData[['survived', 'pclass', 'age', 'sibsp', 'parch', 'fare']]
CategoryData=titanicData[[ 'sex','embarked', 'class', 'who', 'adult_male', 'deck', 'embark_town','alive', 'alone']]
#Lets print the unique values/classification/options present in the CategoryData  columns dataframe.
print(CategoryData['sex'].unique())
print(CategoryData['embarked'].unique())
print(CategoryData['class'].unique())
print(CategoryData['who'].unique())
print(CategoryData['adult_male'].unique())
print(CategoryData['deck'].unique())
print(CategoryData['embark_town'].unique())
print(CategoryData['alive'].unique())
print(CategoryData['alone'].unique())
# NumeriDAta.head(n=10)
# NumeriDAta.describe()
CategoryData.head()
#titanicData.columns
# The easiest way to convert a column into one hot-encoded column
# is by using the get_dummies() method of the Pandas dataframe, as shown below:
#Then you can cast the entire DataFrame to integer type: ie Convert True/False  1/0 using astype(int) method
encoded_Category_Data=pd.get_dummies(CategoryData[['sex','embarked', 'class', 'who',
                                                   'adult_male', 'deck', 'embark_town','alive', 'alone']]).astype(int)
# Also, you can create one hot encoded columns for null values in the
# actual column by passing True as a value for the dummy_na parameter.
encoded_Category_Data=pd.get_dummies(CategoryData[['sex','embarked', 'class', 'who', 'adult_male', 'deck',
                                                   'embark_town','alive', 'alone']],dummy_na=True).astype(int)
#Lets display the actual CategoryData and the one hot encoded version ie encoded_Category_Data in the same dataframe
#using the concat() method from pandas
#Merging the dataframes
mergeddataFrame=pd.concat([CategoryData,encoded_Category_Data],axis=1)
#mergeddataFrame.info()
CompleteDataFrame1=pd.concat([NumeriDAta,mergeddataFrame],axis=1)
CompleteDataFrame=pd.concat([NumeriDAta,encoded_Category_Data],axis=1)
#CompleteDataFrame.info()
#CompleteDataFrame.describe()
#CompleteDataFrame.isnull().mean()
#Filling NaN value in the age column with the age mode
#CompleteDataFrame=CompleteDataFrame.fillna({"age":CompleteDataFrame['age'].mode()[0]})
CompleteDataFrame['ModeAge']=CompleteDataFrame['age'].fillna(CompleteDataFrame['age'].mode()[0])
plt.rcParams['figure.figsize']=[8,10]
ax=plt.figure().add_subplot(1,1,1)
CompleteDataFrame['age'].plot(kind='kde',ax=ax)
CompleteDataFrame['ModeAge'].plot(kind='kde',ax=ax)
#CompleteDataFrame['sex_female'].plot(kind='kde',ax=ax)
#print(CompleteDataFrame.columns)
#print(CompleteDataFrame.head())
sns.set_style('darkgrid')
lines,lables=ax.get_legend_handles_labels()
ax.legend(lines,lables,loc="best")
#CompleteDataFrame.isnull().mean()



```

```python
from ast import Assign
# Handling Outliers
# Outliers are the values that are too far from the rest of the
# observations in the columns. For instance, if the weight of most of
# the people in the sample varies between 50100 kgs, an
# observation of 500kg will be considered as an outlier since such an observation occurs rarely.
# There are four main techniques to handle outliers:
# 1. You can either totally remove the outliers from the dataset.
# 2. You can treat outliers as missing values and then apply any d
# imputation technique that you studied in chapter 3.
# 3. You can apply discretization techniques to a dataset that will
# include outliers along with other data points at the tail.
# 4. You can cap or censor the outliers and replace them with
# maximum and minimum values that can be found via several
# # techniques.
# Outlier Trimming
# Outlier trimming, as the name suggests, refers to simply removing
# the outliers beyond a certain threshold value.
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
datta=sns.load_dataset('titanic')
datta=datta.fillna({'age':datta['age'].mode()[0]})
#print(datta.head())
#The best way to visualize outliers is using boxplots
# Q1=np.percentile(datta['age'],25)
# Q2=np.percentile(datta['age'],50)
# Q3=np.percentile(datta['age'],75)
# IQR=Q3-Q1

# print(datta['age'].mean())
# print("25th percentile (Q1):",Q1)
# print("50th percentile (Median, Q2):", Q2)
# print("75th percentile (Q3):", Q3)
# print("IQR (Interquartile Range) = Q3  Q1:", IQR)
# print("Lower Box Plot whisker = Q1  1.5  IQR:", Q1-1.5*IQR)#Any value below this is considered as an outlier -it's represented by black dots/circles of the boxplot. This is the lower limit value
# print("Upper Box Plot whisker = Q3 + 1.5  IQR:", Q3+1.5*IQR)#Any value above this is considered as an outlier -it's represented by black dots/circles of the boxplot. This is the upper limit value
#so anything above/below whiskers is considered an outlier
#sns.boxplot(data=datta['age'])
#print(datta['age'].describe())
# To remove outliers, we first need to define the values that will be
# considered outliers. There are various ways to do so. One of the
# most common ways to do so is to find the Inter Quartile Range
# (IQR), multiply it by 1.5, and then subtract it from the first quartile
# value (0.25 quantile) to find the lower limit. To find the upper limit,
# add the product of IQR and 1.5 to the 3
# rd quartile value (0.75
# quantile). IQR can be calculated by subtracting the first quartile
# value from the 3th quartile.
#You can use the percentiles from numpy or the quantile() method as below . Both give the same results
Q1=datta['age'].quantile(0.25)
Q3=datta['age'].quantile(0.75)
IQR=Q3-Q1
lower_age_limit=Q1-1.5*IQR
upper_age_limit=Q3+1.5*IQR
print(lower_age_limit,upper_age_limit)
#1. Create an outlier flag column using np.where() or dataframe_name['column_Name'].between(lower_bound,upper_bound)
# np.where has 3 arguments:
# np.where(condition, value_if the condition is_true, value if the condition is false)
#The below script using the where() method from numy find the rows containing
# the outlier values and Assign them with True value else Fale
#you must wrap each condition in parentheses.ie (your 1st condition)| (other condition)
# np.where(condition)  gives indices where condition is True.
# np.where(condition, x, y)  works like a vectorized if-else.
#datta['age_outlierfilter']=np.where(datta['age']>upper_age_limit | datta['age']<lower_age_limit,True,False) #THIS IS WRONG BECAUSE THE CONDITION IS NOT WRAPPED IN A PARETHESIS

datta['age_outlierfilter'] = np.where((datta['age'] > upper_age_limit) | (datta['age'] < lower_age_limit),
                                      True#Outlier
                                      , False# Not Outlier
                                      )

#print(datta[['age_outlierfilter','age']].head(60))

# ALTERNATIVELY, you can also use .between() to filter non-outliers directly:
#datta['age_outlierfilter']=~datta['age'].between(lower_age_limit,upper_age_limit)
#Here, ~ negates the mask, so values outside the limits ie (lower_age_limit, upper_age_limit) become True (outliers).
#print(datta[['age_outlierfilter','age']].head(60))
#2. Filter out/remove the outliers
# You can remove entire rows where age_outlierfilter is True:
datta_with_no_outliers = datta[datta['age_outlierfilter'] == False]
print(datta_with_no_outliers['age'].head(60))
#datta_with_no_outliers.drop(columns='age_outlierfilter')=datta.drop(labels='age_outlierfilter',axis=1) ie both expression/code deletes the age_outlierfilter column
#print(datta_with_no_outliers.shape,datta.shape)
#sns.boxplot(data=datta_with_no_outliers['age'])

#Outlier Capping using lower and upper bounds
# Outlier capping is a technique where you do not remove outliers(which may reduce data size), but instead limit (cap)
#  their values within a reasonable range (like lower and upper bounds).
# np.where has 3 arguments:
#1. Create an non-outlier dataframe
# np.where(condition, value_if the condition is_true, value if the condition is false)
#The below script using the where() method from numy find the rows containing
datta['age_capped']=np.where(datta['age']<lower_age_limit,lower_age_limit,np.where(datta['age']>upper_age_limit,
                                           upper_age_limit,datta['age']))
#  What happens in the above script:
# If age < lower_age_limit  replaced with lower_age_limit
# If age >  upper_age_limit  replaced with  upper_age_limit
# Otherwise  keep original value
#PLOTTING
plt.rcParams['figure.figsize']=[8,6]
axiss=plt.figure().add_subplot(1,1,1)
sns.boxplot(data=datta['age_capped'],ax=axiss)#No outliers
#sns.boxplot(data=datta['age'],ax=axiss)# Has outliers
# datta['age_capped'].plot(kind='kde',ax=axiss)
# datta['age'].plot(kind='kde',ax=axiss)
lines,lables=axiss.get_legend_handles_labels()
axiss.legend(lines,lables,loc='best')
#print(datta.columns)# both age_capped and age column present
#Now we can delete the age column and use the age_capped column
#datta=datta.drop(columns='age')
#print(datta.columns)#age_capped present and age column deleted




```

```python
# Outlier Capping Using Mean and Std-NOT GOOD TECHNIQUE TO USE MEAN/STD CAUSE OUTLIERS PERSIST
# Instead of using the IQR method, the upper and lower thresholds
# for outliers can be calculated via the mean and standard deviation
# method. To find the upper threshold, the mean of the data is added
# to three times the standard deviation value. Similarly, to find the
# lower threshold, you have to multiply the standard deviation by 3
# and then remove the result from the mean.
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
datta=sns.load_dataset('titanic')
datta=datta.fillna({'age':datta['age'].mode()[0]})
Q1=datta['age'].quantile(0.25)
Q3=datta['age'].quantile(0.75)
IQR=Q3-Q1
lower_age_limit=Q1-1.5*IQR
upper_age_limit=Q3+1.5*IQR
lower_bound=datta['age'].mean()-3*datta['age'].std()
upper_bound=datta['age'].mean()+3*datta['age'].std()
datta['capped_age']=np.where(datta['age']<lower_age_limit,lower_age_limit,np.where(datta['age']>upper_age_limit,
                                           upper_age_limit,datta['age'])) #SUITABLE for replacing OUTLIERS
#Since it elimates all outliers in the data

# datta['capped_age']=np.where(datta['age']<lower_bound,lower_bound,np.where(datta['age']>upper_bound,
#                                            upper_bound,datta['age']))#NOT SUITABLE for replacing OUTLIERS
# datta['capped_age']=np.where(datta['age']<lower_bound,datta['age'].mode()[0],np.where(datta['age']>upper_bound,
#                                            datta['age'].mode()[0],datta['age']))#NOT SUITABLE for replacing OUTLIERS
plt.rcParams['figure.figsize']=[8,6]
axiss=plt.figure().add_subplot(1,1,1)
sns.boxplot(data=datta['capped_age'],ax=axiss)#No outliers
#print(lower_bound,upper_bound)
```

```python
# Solving Regression Problems in Machine
# Learning Using Sklearn Library
# Machine learning is a branch of artificial intelligence that enables
# computer programs to automatically learn and improve from
# experience. Machine learning algorithms learn from datasets and
# then, based on the patterns identified from the datasets, make
# predictions on unseen data.
# Machine learning algorithms can be broadly categorized into two
# types: Supervised learning and unsupervised learning algorithms.
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
#print(sns.get_dataset_names())
## Loading dataset
data=sns.load_dataset('tips')
data.head()
# Dividing Data into Features and Labels
# in any machine learning workflow, we divide the dataset into:
# Features (X)  the input/independent variables used to predict.
# Labels (y / target)  the dependent/variable we want to predict.
# Why Divide Data?
#  Features are used to train the model.
#  Labels tell the model what the correct answer is.
#  This separation is necessary for training, validation, and testing the model.
X=data.drop(['tip'],axis=1)#removes the variable we want to predict.
y=data['tip']#stores variable we want to predict.
#dividing features into categorical and numerical
X_Categorical=X.drop(['total_bill','size'], axis=1)
X_Numerical=X[['total_bill','size']]
# Converting Categorical Data to Numbers
# Machine learning algorithms can only work with numbers.
# Therefore, it is important to convert categorical data into a numeric format.
# One of the most common approaches to convert a categorical
# column to a numeric one is via one-hot encoding.
#In one-hot encoding, for every unique value in the original columns, a new column is created.
X_Categorical=pd.get_dummies(data=X_Categorical,drop_first=True).astype(int)
# The final step is to join the numerical columns with the one-hot
# encoded columns. To do so, you can use the concat() function
# from the Pandas library, as shown below:
X=pd.concat([X_Categorical,X_Numerical],axis=1)
#print(X.head(30))
# After a machine learning algorithm has been trained, it needs to be evaluated to see how well it performs on
# unseen data. Therefore,we divide the dataset into two sets, i.e., train set and test set. The
# dataset is trained via the train set and evaluated on the test set. To split the data into training and test sets,
#you can use the train_test_split() function from the Sklearn library, as shown
# below. The following script divides the data into 80 percent train set and 20 percent test set.
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.8,random_state=0)
# What does random_state 0 mean?
# The random state hyperparameter in the train_test_split() function controls the shuffling process.
#  With random_state=None, we get different train and test sets across other executions and the shuffling process is
#  out of control. With random_state=0, we get the same train and test sets across different executions.
# Why use random_state=0 (or any number)?
# Reproducibility: You can reliably reproduce the exact same results (e.g., the same data split) each time you run your code.
# Consistency: Ensures that your model training and testing are performed on the exact same data partitions, regardless of how many times the code is executed.
# Debugging: Helps in debugging by providing a consistent data partition to trace problems and verify results.
# Fair Comparison: Allows for fair comparison of different models or hyperparameters, as the data splits remain constant.
#print(X_train.shape,X_test.shape,y_train.shape,y_test.shape)
#Data scaling/normalization
X_train=StandardScaler().fit(X_train).transform(X_train)
X_test=StandardScaler().fit(X_test).transform(X_test)


```

```python
#Linear Regression
import pandas as pd
import seaborn as sns
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
#Loading dataset
data=sns.load_dataset('tips')
# Dividing Data into Features and Labels/Target Variable
X=data.drop(['tip'],axis=1)#removes the variable we want to predict.
y=data['tip']#stores variable we want to predict.
#print(y.info())#y is a pandas series
#sns.boxplot(data=y)#CHECK for outliers
#Capping/REMOVING OUTLIERS IN y series- Series has one column so we don't need a for in loop
#but we need where() method from numpy
Q1=y.quantile(0.25)
Q3=y.quantile(0.75)
IQR=Q3-Q1
lower_limit=Q1-1.5*IQR
upper_limit=Q3+1.5*IQR
y=np.where(y<lower_limit,lower_limit,
        np.where(y>upper_limit,upper_limit,y))
#print(y.dtype) #float array
#recreate a pandas series of y and store it y variable
y=pd.Series(y)
print(y.info())
sns.boxplot(data=y)#CHECK for outliers-No Outlier anymore
#sns.boxplot(data=X['total_bill'])#CHECK for outliers in the feature column-Outliers present
#dividing features into categorical and numerical
data_numeric=X.select_dtypes(include=['number'])
data_categorical=X.select_dtypes(exclude=['number'])
#Converting/Encoding Categorical DataFrame using one hot encoding
data_categorical=pd.get_dummies(data=data_categorical,drop_first=True).astype(int)
#Merging the Dataset
X_Merged_Data=pd.concat([data_numeric,data_categorical],axis=1)
# X_Merged_Data.head()
#sns.boxplot(data=X_Merged_Data['total_bill'])#CHECK for outliers in the total_bill feature column
#REMOVING OUTLIERS IN FEATURES using for in loop and where() method from numpy
for column in X_Merged_Data.columns:
  Q1=X_Merged_Data[column].quantile(0.25)
  Q3=X_Merged_Data[column].quantile(0.75)
  IQR=Q3-Q1
  lower_limit=Q1-1.5*IQR
  upper_limit=Q3+1.5*IQR
  X_Merged_Data[column]=np.where(X_Merged_Data[column]<lower_limit,lower_limit,
        np.where(X_Merged_Data[column]>upper_limit,upper_limit,X_Merged_Data[column]))
#sns.boxplot(data=X_Merged_Data['total_bill'])# no more outliers
#dividing the dataset into two sets, i.e., train set and test set.
X_train,X_test,y_train,y_test=train_test_split(X_Merged_Data,y,random_state=0,train_size=0.8)
#print(X_train.shape,X_test.shape,y_train.shape,y_test.shape)
#print(X_train.head(),X_test.head(),y_train.head(),y_test.head())
#Data scaling/normalization
X_test=StandardScaler().fit(X_test).transform(X_test)
X_train=StandardScaler().fit(X_train).transform(X_train)
# Single Output Regression Problems
# In this section, you will solve regression problems with one possible
# output. In a regression problem with a single output, you have to
# predict a single continuous value, such as the price of a house, the
# blood pressure of a patient, or the marks of a student. In multioutput regression problems,
# there is more than one continuous value in the output.
# Linear Regression
# Linear regression is a linear model that assumes a linear
# relationship between inputs and outputs and minimizes the cost of
# error between the predicted and actual output using functions like
# mean absolute error.
#TRAINING THE MODEL
model=LinearRegression().fit(X_train,y_train)
#MODEL PREDICTION ON UNSEEN DATA
model_to_PredictValue_of_y=model.predict(X_test)
#print(model_to_PredictValue_of_y)
model_predicted_Dataframe=pd.DataFrame(data=model_to_PredictValue_of_y,columns=['Predicted_Tip'])
actual_value_tip=pd.DataFrame(data=y_test.reset_index(drop=True),columns=['Actual_Tip'])
merged_y_pred_y_test=pd.concat([actual_value_tip,model_predicted_Dataframe],axis=1)
# print(merged_y_pred_y_test)
merged_y_pred_y_test.plot(kind='kde')

#EVALUATING MODEL PERFORMANCE
# Once you have trained a model and have made predictions on the test set, the next step is to know how well your model has
# performed for making predictions on the unknown test set. There are various metrics to check that. However, mean absolute error,
# mean squared error, and root mean squared error are three of the most common metrics.
print("mean_absolute_error=",metrics.mean_absolute_error(y_test,model_to_PredictValue_of_y))
print("mean_absolute_percentage_error=",metrics.mean_absolute_percentage_error(y_test,model_to_PredictValue_of_y))
print("mean_squared_error:",metrics.mean_squared_error(y_test,model_to_PredictValue_of_y))
# By looking at the mean absolute error, it can be
# concluded that, on average, there is an error of 0.57. for
# predictions, which means that on average, the predicted tip values
# are 0.57$ more or less than the actual tip values.

```

```python
# Dimensionality Reduction with PCA and LDA
# using Sklearn
# Dimensionality reduction refers to reducing the number of features
# in a dataset in such a way that the overall performance of the
# algorithms trained on the dataset is minimally affected.
# There are three main approaches used for dimensionalityreduction:
# Principal Component Analysis (PCA), Linear
# Discriminant Analysis (LDA), and Singular Value Decomposition
# # (SVD).
# Principal Component Analysis
# Principal component analysis is an unsupervised dimensionality
# reduction technique that doesnt depend on the labels of a dataset.
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
#from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
#Loading data
data=sns.load_dataset('iris')
#sns.boxplot(data=data['petal_length'])
#sns.boxplot(data=data['petal_width'])
# print(data.info)
# print(data.dtypes)
#print(data.head(3))
#Dividing features and lables
X=data.drop(['species'],axis=1)
y=data['species']
#print(y.unique())
#Label Encoding on lables using lablencoder class from sklearn
y=LabelEncoder().fit(y).transform(y)
#print(y.dtype)
#print(y)
#Scaling the features dataset
X_Scaled=StandardScaler().fit(X).transform(X)
#print(X_Scaled)
#Converting the Scaled features to a pandas dataframe
X=pd.DataFrame(data=X_Scaled,columns=X.columns)
#print(X.describe())
# Before we apply PCA on a dataset, we will divide it into the training and test sets.
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=1,train_size=0.8)

# To apply PCA via Sklearn, all you have to do is import the PCA class
# from the Sklearn.decomposition module. Next, to apply PCA to the
# training set, pass the training set to the fit_tansform() method of
# the PCA class object. To apply PCA on the test set, pass the test
# set to the transform() method of the PCA class object.
pca=PCA(n_components=2)
#Training the PCA model to the training set
X_train=pca.fit_transform(X_train)
#print(X_train)
#Making prediction on the test data
X_test=pca.fit_transform(X_test)
# #print(X_test)
# Once you have applied PCA on a dataset, you can use the
# explained_variance_ratio_ feature to print the variance caused by
# all the features in the dataset.

#Printing the variance ratio
variance_ratio=pca.explained_variance_ratio_
#print(variance_ratio) #[0.72042188 0.25549267 0.01659378 0.00749168]
# The output above shows that 72.042188 percent of the variance in the
# dataset is caused by the first principal component, while 25.549267
# percent of the variance is caused by the second principal
# component.
# Lets now select the two principal components that collectively
# caused variance of 72.042188 % + 25.549267% = 97.589%.

#TRaining the logistic regression model
lg_Model=LogisticRegression().fit(X_train,y_train)
#Model Prediction
y_predict=lg_Model.predict(X_test)
print(y_predict.shape)
y_predict=pd.DataFrame(data=y_predict,columns=['Predicted_species'])
y_actual =pd.DataFrame(data=y_test,columns=['Actual_species'])
#merging the DataFrames
merged_y_pred_y_test=pd.concat([y_actual,y_predict],axis=1)
#print(merged_y_pred_y_test.head(20))
#Model Evaluation
print("Accuracy Score:",accuracy_score(y_test,y_predict))
# The output shows that even with two features, the accuracy for
# correctly predicting the label for the iris plant is 93.33333333333333.
plt.scatter(X_test[:,0],X_test[:,1],c=y_test,cmap='rainbow')
plt.show()
```
